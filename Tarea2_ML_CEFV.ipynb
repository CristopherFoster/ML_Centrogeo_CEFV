```python
import numpy as np
import plotly.graph_objects as go
```

```python
# Se define una matriz sigma positiva
# Se puede alterar los valores siempore y cuando sea semidefinida positiva
Sigma = np.array([
    [4, 1, 2],
    [1, 3, 1],
    [2, 1, 5]
])

# Se verifica por si las dudas que Sigma sea semidefinida positiva
eigenvalues = np.linalg.eigvals(Sigma)
if np.all(eigenvalues >= 0):
    print("Sigma es semidefinida positiva")
else:
    print("Sigma no es semidefinida positiva")

# Se define la función de la norma
#$|x\|_{\Sigma}^2 = x^T \Sigma $
def norm_sigma(x, Sigma):
    return np.sqrt(np.dot(x.T, np.dot(Sigma, x)))

# Se crean un conjunto de vectores para evaluar la norma
vectors = [
    np.array([1, 0, 0]),
    np.array([0, 1, 0]),
    np.array([0, 0, 1]),
    np.array([1, 1, 0]),
    np.array([1, 0, 1]),
    np.array([0, 1, 1]),
    np.array([1, 1, 1]),
    np.array([-1, 1, 1]),
    np.array([2, -1, 0])
]

# Se calculó la Norma
#Evaluando la funcion
norms = [norm_sigma(v, Sigma) for v in vectors]

# El comportamiento de la norma se visualizó como puntos en una esfera, por lo que definimos la esfera
phi = np.linspace(0, np.pi, 100)
theta = np.linspace(0, 2 * np.pi, 100)
phi, theta = np.meshgrid(phi, theta)
x = np.sin(phi) * np.cos(theta)
y = np.sin(phi) * np.sin(theta)
z = np.cos(phi)

# Se calculó la norma en cada punto de la esfera
norm_values = np.array([norm_sigma(np.array([x[i, j], y[i, j], z[i, j]]), Sigma) for i in range(x.shape[0]) for j in range(x.shape[1])])
norm_values = norm_values.reshape(x.shape)

#Finalmente representamos la superficie como:
surface = go.Surface(x=x, y=y, z=z, surfacecolor=norm_values, colorscale='Viridis', colorbar=dict(title='Norma'))

# Se visualizan los vectores
vectors_trace = [go.Scatter3d(
    x=[0, v[0]],
    y=[0, v[1]],
    z=[0, v[2]],
    mode='lines+markers',
    line=dict(color='red', width=5),
    marker=dict(size=4),
    name=f"{v}, Norma: {n:.2f}"
) for v, n in zip(vectors, norms)]

# Configuración del layout
layout = go.Layout(
    title="Vectores y superficie de norma con respecto a Sigma",
    scene=dict(
        xaxis=dict(range=[-2, 2]),
        yaxis=dict(range=[-2, 2]),
        zaxis=dict(range=[-2, 2])
    ),
    legend=dict(x=0.02, y=0.95)  # Ajuste de la posición de la leyenda
)

# Crear la figura
fig = go.Figure(data=[surface] + vectors_trace, layout=layout)

# Mostrar la figura
fig.show()
```

```python
# Sigma simétrico positivo e 2D
Sigma = np.array([
    [4, 1],
    [1, 3]
])

# Verificar Sigma semidefinida positiva
eigenvalues = np.linalg.eigvals(Sigma)
if np.all(eigenvalues >= 0):
    print("Sigma es semidefinida positiva")
else:
    print("Sigma no es semidefinida positiva")

# Definir la función de norma
def norm_sigma(x, Sigma):
    return np.sqrt(np.dot(x.T, np.dot(Sigma, x)))

# Se define a un conjunto de vectores para evaluar la norma
vectors = [
    np.array([1, 0]),
    np.array([0, 1]),
    np.array([1, 1]),
    np.array([-1, 1]),
    np.array([2, -1])
]

# Calculo de las normas
#Aplicando la función sigma en un ciclo for horizontal, para todos los vectores
norms = [norm_sigma(v, Sigma) for v in vectors]

# Crear una malla de puntos para evaluar la norma
x_range = np.linspace(-3, 3, 100)
y_range = np.linspace(-3, 3, 100)
x_mesh, y_mesh = np.meshgrid(x_range, y_range)

# Calcular la norma en cada punto de la malla
norm_values = np.array([norm_sigma(np.array([x, y]), Sigma) for x, y in zip(np.ravel(x_mesh), np.ravel(y_mesh))])
norm_values = norm_values.reshape(x_mesh.shape)

# Crear el gráfico de contorno con Plotly
contour = go.Contour(
    x=x_range,
    y=y_range,
    z=norm_values,
    colorscale='Viridis',
    colorbar=dict(title='Norma')
)

# Crear vectores con Plotly
vectors_trace = [go.Scatter(
    x=[0, v[0]],
    y=[0, v[1]],
    mode='lines+markers',
    line=dict(color='red', width=5),
    marker=dict(size=4),
    name=f"Vector {v}, $\\|x\\|_{{\\Sigma}} = {n:.2f}$"
) for v, n in zip(vectors, norms)]

# Configurar el layout
layout = go.Layout(
    title="Vectores y gráfico de contorno de la norma con respecto a Sigma",
    xaxis=dict(range=[-3, 3]),
    yaxis=dict(range=[-3, 3], scaleanchor='x', scaleratio=1),
    width=600,
    height=600,
    legend=dict(x=0.02, y=0.95)  # Ajuste de la posición de la leyenda
)

# Crear la figura
fig = go.Figure(data=[contour] + vectors_trace, layout=layout)

# Mostrar la figura
fig.show()
```
